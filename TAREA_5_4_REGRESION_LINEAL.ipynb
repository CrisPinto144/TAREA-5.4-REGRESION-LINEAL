```python
#   Codigo que implementa la regresion lineal
#   para ajustar un conjunto de datos
#
#
#           Autor:
#   Cristian Yuseff Pinto Villanueva
#   cristian.pino984@mail.com
#   Version 1.0 : 28/04/2025
#

import numpy as np
import matplotlib.pyplot as plt

# Datos de ejemplo
x = np.array([0, 2, 4, 6, 8])
y = np.array([100, 92, 85, 78, 71])

# Cálculo de los coeficientes
n = len(x)
sum_x = np.sum(x)
sum_y = np.sum(y)
sum_xy = np.sum(x * y)
sum_x2 = np.sum(x**2)

# Fórmulas de regresión lineal
b = (n * sum_xy - sum_x * sum_y) / (n * sum_x2 - sum_x**2)
a = (sum_y - b * sum_x) / n

print(f"Coeficientes de la regresión:")
print(f"a (intercepto) = {a:.4f}")
print(f"b (pendiente) = {b:.4f}")

# Predicción usando el modelo
y_pred = a + b * x

# Gráfica
plt.figure(figsize=(8,6))
plt.plot(x, y, 'o', label='Datos')
plt.plot(x, y_pred, '-', label=f'Ajuste lineal: y = {a:.2f} + {b:.2f}x', color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Regresión Lineal')
plt.legend()
plt.grid(True)
plt.savefig('regresion_lineal.png', dpi=300)  # Guarda la figura
plt.show()

```

```python
#EJERCICIO2

import numpy as np
import matplotlib.pyplot as plt

# Datos del ejercicio: Posición (cm) y Temperatura (°C)
x = np.array([0, 2, 4, 6, 8])
y = np.array([100, 92, 85, 78, 71])

# Cálculo de los coeficientes
n = len(x)
sum_x = np.sum(x)
sum_y = np.sum(y)
sum_xy = np.sum(x * y)
sum_x2 = np.sum(x**2)

# Fórmulas de regresión lineal: y = a + bx
b = (n * sum_xy - sum_x * sum_y) / (n * sum_x2 - sum_x**2)
a = (sum_y - b * sum_x) / n

print("Coeficientes de la regresión lineal:")
print(f"a (intercepto) = {a:.4f}")
print(f"b (pendiente)  = {b:.4f}")

# Estimación de la temperatura en x = 5 cm
x_est = 5
y_est = a + b * x_est
print(f"\nEstimación de la temperatura en x = {x_est} cm: {y_est:.2f} °C")

# Predicción para los datos
y_pred = a + b * x

# Gráfica
plt.figure(figsize=(8,6))
plt.plot(x, y, 'o', label='Datos experimentales')
plt.plot(x, y_pred, '-', color='red', label=f'Regresión lineal: y = {a:.2f} + {b:.2f}x')
plt.axvline(x_est, color='gray', linestyle='--', label=f'Estimación en x = {x_est} cm')
plt.axhline(y_est, color='blue', linestyle='--')
plt.plot(x_est, y_est, 'bo')  # Punto estimado
plt.xlabel('Posición (cm)')
plt.ylabel('Temperatura (°C)')
plt.title('Modelo de Regresión Lineal - Transferencia de Calor')
plt.legend()
plt.grid(True)
plt.show()
```
